```python
# Import necessary libraries
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
import matplotlib.pyplot as plt

# Load the Iris dataset
iris = datasets.load_iris()
X = iris.data
y = iris.target

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Standardize the features (important for methods like LDA)
scaler = StandardScaler()
X_train_std = scaler.fit_transform(X_train)
X_test_std = scaler.transform(X_test)

# Apply Linear Discriminant Analysis
lda = LinearDiscriminantAnalysis(n_components=2)
X_train_lda = lda.fit_transform(X_train_std, y_train)
X_test_lda = lda.transform(X_test_std)

# Plot the transformed data
plt.figure(figsize=(8, 6))
for label in range(3):
    plt.scatter(X_train_lda[y_train == label, 0], X_train_lda[y_train == label, 1], label=f'Class {label}')

plt.title('Linear Discriminant Analysis of Iris dataset')
plt.xlabel('LD1')
plt.ylabel('LD2')
plt.legend()
plt.show()
```

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.datasets import load_iris
import pandas as pd

# Load example dataset (Iris dataset)
iris = load_iris()
X = iris.data
y = iris.target

# Create a DataFrame for visualization purposes
df = pd.DataFrame(data=X, columns=[f"Feature_{i+1}" for i in range(X.shape[1])])
df["Target"] = y

# Initialize LDA and fit the model
lda = LinearDiscriminantAnalysis(n_components=1)
X_lda = lda.fit_transform(X, y)

# Display the weights assigned to each feature by LDA
feature_weights = pd.DataFrame(lda.coef_.T, columns=[f"LDA_Weight_Class_{i}" for i in range(lda.coef_.shape[0])], index=df.columns[:-1])
print("Feature weights assigned by LDA:\n", feature_weights)

# Transform the original data to the reduced feature space
X_lda = lda.transform(X)

# Display the transformed data
df_lda = pd.DataFrame(data=X_lda, columns=["LDA_Component"])
df_lda["Target"] = y

print("\nTransformed data using LDA:\n", df_lda.head())
```

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.datasets import load_breast_cancer
import pandas as pd

# Load breast cancer dataset
cancer = load_breast_cancer()
X = cancer.data
y = cancer.target

# Create a DataFrame for visualization purposes
df = pd.DataFrame(data=X, columns=[f"Factor_{i+1}" for i in range(X.shape[1])])
df["Target"] = y

# Initialize LDA and fit the model
lda = LinearDiscriminantAnalysis(n_components=1)
X_lda = lda.fit_transform(X, y)

# Display the weights assigned to each feature by LDA
feature_weights = pd.DataFrame(lda.coef_.T, columns=["LDA_Weight"], index=df.columns[:-1])
print("Feature weights assigned by LDA:\n", feature_weights)  # The sign of the weight indicates whether the feature tends to have a higher value for one class relative to the other.

# Transform the original data to the reduced feature space
X_lda = lda.transform(X)

# Display the transformed data
df_lda = pd.DataFrame(data=X_lda, columns=["LDA_Component"])
df_lda["Target"] = y

print("\nTransformed data using LDA:\n", df_lda.head())
```
