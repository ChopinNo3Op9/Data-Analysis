```python
# Convert 'Last Update' column to a datetime object
df['Last Update'] = pd.to_datetime(df['Last Update'])

# Set the 'Last Update' column as the DataFrame index
df.set_index('Last Update', inplace=True)

# Plot YTD Rtn trends for each index
plt.figure(figsize=(12, 8))
for index_name in df['Index Name'].unique():
    index_data = df[df['Index Name'] == index_name]
    plt.plot(index_data.index, index_data['YTD Rtn'], label=index_name)

plt.title('YTD Return Trends for Different Indices')
plt.xlabel('Date')
plt.ylabel('YTD Return')
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
plt.grid(True)
plt.xticks(rotation=45)
plt.show()
```
