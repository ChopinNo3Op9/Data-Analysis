```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Sample data
x = [1, 2, 3, 4, 5]
y = [2, 3, 10, 7, 11]

plt.figure(figsize=(8, 4))  # Set the figure size
plt.plot(x, y, marker='o', linestyle='-', color='b', label='Line Chart')
plt.title('Line Chart Example')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid(True)  # Add grid lines
plt.legend()  # Add legend
plt.show()
```

```python
# Scatter Plot
plt.figure(figsize=(8, 4))
plt.scatter(x, y, color='r', marker='o', s=100, label='Scatter Plot')
plt.title('Scatter Plot Example')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid(True)
plt.legend()
plt.show()
```

Bar charts are used to compare and display the values of different categories or discrete data points.

```python
categories = ['Category 1', 'Category 2', 'Category 3', 'Category 4', 'Category 5']
values = [15, 20, 10, 30, 25]

plt.figure(figsize=(8, 4))
plt.bar(categories, values, color='g', label='Bar Chart')
plt.title('Bar Chart Example')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.grid(axis='y')  # Add grid lines only on the y-axis
plt.legend()
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.show()
```

Histograms are used to represent the distribution of continuous or numerical data. They are used to show the frequency or count of data points within specific intervals or bins.

```python
data = [2, 2, 2, 3, 3, 3, 4, 4, 5, 5, 5, 5, 5, 5, 5, 7, 7, 7, 7, 7, 7, 7, 7, 11, 11, 11]
plt.figure(figsize=(8, 4))
plt.hist(data, bins=3, color='purple', edgecolor='black', alpha=0.1, label='Histogram') # alpha 0.1: 90% opaque and 10% transparent
plt.title('Histogram Example')
plt.xlabel('Bins')
plt.ylabel('Frequency')
plt.grid(axis='y')
plt.legend()
plt.show()
```

```python
# Create a sample data frame
data = [[1, 2, 3],
        [4, 5, 6],
        [7, 8, 9]]

# Create a heatmap
sns.heatmap(data, annot=True, cmap='coolwarm', fmt='d')
plt.show()
```

```python
# Create sample data
x = np.linspace(0, 2*np.pi, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)  # Add y3
y4 = np.exp(-x)  # Add y4

# Create subplots
plt.figure(figsize=(14, 4))

plt.subplot(1, 4, 1)  # 1 row, 4 columns, subplot 1
plt.plot(x, y1)
plt.title('Sine Wave')

plt.subplot(1, 4, 2)  # 1 row, 4 columns, subplot 2
plt.plot(x, y2)
plt.title('Cosine Wave')

plt.subplot(1, 4, 3)  # 1 row, 4 columns, subplot 3
plt.plot(x, y3)
plt.title('Tangent Wave')

plt.subplot(1, 4, 4)  # 1 row, 4 columns, subplot 4
plt.plot(x, y4)
plt.title('Exponential Decay')

plt.tight_layout()
plt.show()
```
